モンテカルロ法の推定誤差...乱数を使用したモンテカルロ法は、実行のたびに推定値が変わるという特徴がある。この推定誤差は以下の3つに大別される。$\\$

①サンプル数による誤差...大数の法則に基づいて、真の平均に近づいていく。$\\$
②離散化による誤差...様々な離散化手法があるが、それにより誤差が生ずる。（正確な離散化を行うと、モンテカルロ推定値は不偏推定量となる。）$\\$
③丸め誤差...２進法の世界であるコンピュータが、限られたメモリーでメモリーで10進法を表現する際に起こる誤差。

①サンプル数による誤差...大数の法則に基づいて、真の平均に近づいていく。また、モンテカルロ推定値は（離散化が正確な場合）不偏推定量となるため、たとえサンプル数が少なく個々の推定値が大きくぶれている場合でも標本平均は真の平均値にきわめて近い値をとる。

```python
# サンプル数100の推定値の分布
N =100
# 不偏推定量であることを確かめるため、正確な離散化式を使う
from common.option_price import BS_Monte_Carlo

s0 = 100
K = 100
r = 0.05
T = 20
mu = r
sigma = 0.3
n = 1
dt = T/n
BS_Call = BS_Monte_Carlo(s0, mu, sigma, dt, N)

Call_sum = 0
Call_vec = []
for i in range(N):
    Call = BS_Call.MC_Call(s0, K, r, T, mu, sigma, dt, N)
    Call_sum = Call_sum + Call
    Call_vec.append(Call)
Call_mean = Call_sum / N

print(Call_mean)
Call_vec = np.array(Call_vec)

# コール価格の分布
data1 = Call_vec

# 平均値と標準偏差の計算
mean1 = np.mean(data1)
std1 = np.std(data1)

# 信頼区間の計算
confidence = 0.99
n = len(data1)
margin = stats.t.ppf((1 + confidence) / 2, n - 1) * std1 / np.sqrt(n)
lower_bound = mean1 - margin
upper_bound = mean1 + margin

# グラフの描画
plt.figure()
plt.hist(data1, bins=30, density=True, alpha=0.7, label='Histogram')
# x = np.linspace(min(data2), max(data2), 100)
x = np.linspace(9, 135, 100)
plt.plot(x, stats.norm.pdf(x, mean1, std1), color='red', label='Normal Distribution')
plt.axvline(lower_bound, color='blue', linestyle='--', label='Lower Bound (99% Confidence)')
plt.axvline(upper_bound, color='blue', linestyle='--', label='Upper Bound (99% Confidence)')
plt.xlabel('Call Price')
plt.ylabel('Density')
plt.title('Sample Number 100')
plt.legend()
plt.show()
```

（注）上では、繰り返しサンプリングした価格の標本平均と標本標準偏差を母集団のパラメータとみなして、コール価格の信頼区間を推定しているが、サンプル数が少ないとこれらの値ももちろんブレる。よって、一般に信頼区間の推定には注意が必要である。（今の場合は、真の平均である母集団の平均（理論コール価格）がわかっているので、それを用いるとより精度の高い信頼区間の推定も可能ではある。）

推定誤差の推定...ランダム変量$V$から抽出した推定量を$\tilde{v}$とする（例えばモンテカルロBSコール価格）。このとき、$N$個のサンプルによる推定値（標本平均）は以下のようになる。
$$\tilde{v}_N=\cfrac{1}{N}\displaystyle{\sum_{i=1}^N}v_i$$
中心極限定理より、$N$が十分大きいとき、正規分布に分布収束する。$V$の真の分散を$\sigma^2$とすると
$$\tilde{v}_N\rightarrow\phi\left(\mu, \cfrac{\sigma}{\sqrt{N}}\right)$$


推定精度の向上法...モンテカルロ法において、推定精度と計算時間はトレードオフの関係にある。ここでは、計算時間を保ちながら推定精度を上げる、推定精度を保ちながら計算時間を減らす、分散減少法の1つである負相関変量法をみる。

負相関変量法（Antithetic Variable Technique）...推定精度を向上させるテクニックの1つで、サンプルされた正規乱数に対して、符号を逆にしたもう1つの正規乱数をサンプルすることで推定精度の向上を図る。1回の試行で、所与の正規乱数$\varepsilon$とその負の乱数$(-\varepsilon)$により2つのペイオフを得る。

$$\Delta S=\mu S\Delta t+\sigma S\varepsilon\sqrt{\Delta t}$$
$$\Delta S=\mu S\Delta t+\sigma S(-\varepsilon)\sqrt{\Delta t}$$

```python
# 負相関変量法によるパス
dt = 0.01
s = 1000
mu = 0.5
sigma = 0.3
T = 3
jvec = []
svec_p = []
svec_n = []
dsvec_p = []
dsvec_n = []

for j in range(T*100):
    eps = np.random.normal(0,1,1)
    ds_p = mu * s * dt + sigma * s * eps * np.sqrt(dt)
    ds_n = mu * s * dt - sigma * s * eps * np.sqrt(dt)
    s_p = s + ds_p
    s_n = s + ds_n
    jvec.append(j*dt)
    svec_p.append(s_p)
    svec_n.append(s_n)
    dsvec_p.append(ds_p)
    dsvec_n.append(ds_n)

plt.plot(jvec, svec_p)
plt.plot(jvec, svec_n)
plt.show()
```

```python
# 正確な離散化よるモンテカルロ法（負相関）
class BS_Monte_Carlo:
    def __init__(self, s0, mu, sigma, dt, N):
        self.s0 = s0
        self.mu = mu
        self.sigma = sigma
        self.dt = dt
        self.N = N
    
    def Price(self, s0, mu, sigma, dt):
        eps = np.random.normal(0,1,1)
        s = s0 * np.exp((mu - (sigma**2) / 2 ) * dt + sigma * eps * np.sqrt(dt))
        return s
    
    def MC_Call(self, s0, K, r, T, mu, sigma, dt, N):
        Payoff = 0
        Payoff2 = 0
        for i in range(N):
            s = self.Price(s0, mu, sigma, dt)
            dPayoff = max(s-K, 0)
            Payoff = Payoff + dPayoff
            # 負相関変量法
            s2 = self.Price(s0, mu, -sigma, dt)
            dPayoff2 = max(s2-K, 0)
            Payoff2 = Payoff2 + dPayoff2
        mean_Payoff = (Payoff + Payoff2) / (2*N)  # 2つのペイオフ
        MC_call = mean_Payoff * np.exp(-r * T)
        return MC_call
```

```python
s0 = 100
K = 100
r = 0.05
T = 1
mu = r
sigma = 0.3
n = 1
dt = T/n
N =1000

BS = BS_Monte_Carlo(s0, mu, sigma, dt, N)

C = BS.MC_Call(s0, K, r, T, mu, sigma, dt, N)
print(f'MC_Call = {C}')
```

②離散化による誤差...Euler法は簡便な反面、連続関数を微小区間に分けて近似推定するため、常に推定誤差を伴う。

$$S_{t+\Delta t}=S_t+r S_t\Delta t+\sigma S_t\varepsilon\sqrt{\Delta t}$$

```python
# Euler法によるモンテカルロ法
class BS_Monte_Carlo_Euler:
    def __init__(self, s0, mu, sigma, dt, n, N):
        self.s0 = s0
        self.mu = mu
        self.sigma = sigma
        self.dt = dt
        self.n = n  # n = T /dt
        self.N = N
    
    def Price_Euler(self, s0, mu, sigma, dt, n):
        s = s0
        for i in range(n):
            eps = np.random.normal(0,1,1)
            ds = mu * s * dt + sigma * s * eps * np.sqrt(dt)
            s = s + ds
        return s
    
    def MC_Call_Euler(self, s0, K, r, T, mu, sigma, dt, n, N):
        Payoff = 0
        dPayoff_vec = []
        for i in range(N):
            s = self.Price_Euler(s0, mu, sigma, dt, n)
            dPayoff = max(s-K, 0)
            # dPayoff_vec.append(dPayoff)
            dPayoff_vec = np.hstack([dPayoff_vec, dPayoff])
        # 平均値と標準偏差の計算
        mean_Payoff = np.mean(dPayoff_vec) * np.exp(-r * T)
        std_Payoff = np.std(dPayoff_vec) * np.exp(-r * T)**2

        return mean_Payoff, std_Payoff
```

```python
s0 = 100
K = 100
r = 0.05
T = 20
mu = r
sigma = 0.3
n = 50
dt = T/n
N =100

BS = BS_Monte_Carlo_Euler(s0, mu, sigma, dt, n, N)

m,s= BS.MC_Call_Euler(s0, K, r, T, mu, sigma, dt, n, N)
print(f'MC_Call = {C}')
print(m)
print(s)
# print(v)
# print(len(v))
```
